 2. Explain the concept of function arguments and parameters in Python.

   Ans:
   
   Parameters: These are the variables defined in the function's signature, within the parentheses, during the function's definition. They act as placeholders for the values that the function expects to receive when it is called. Parameters specify the type and number of inputs a function is designed to handle.
         
        def greet(name, age): # 'name' and 'age' are parameters
            print(f"Hello, {name}! You are {age} years old.")

   Arguments: These are the actual values or data passed to the function when it is called. Arguments provide the concrete data that the function will operate on, filling the roles defined by the parameters.

        greet("Alice", 30) # "Alice" and 30 are arguments

 4. What is the purpose of the `return` statement in a Python function?

   Ans:
   The return statement in a Python function serves two primary purposes:

 * Exiting the Function:
When a return statement is encountered during the execution of a function, it immediately terminates the function's execution. No further code within that function, after the return statement, will be executed. Control is then passed back to the point in the program where the function was called.
 * Returning a Value:
The return statement can optionally be followed by an expression, which represents the value the function will send back to the caller. This returned value can be of any Python object type, including numbers, strings, lists, dictionaries, or even other functions. If no value is specified after return, or if the return statement is omitted entirely, the function implicitly returns None.

  E.g.
     
       def add(a, b):
       return a + b
       
       result = add(5, 3)
       print("The sum is:", result)

 6. Explain the concept of generators in Python and how they are defined.

   Ans:

   Generators are special functions that yield values one at a time, instead of returning them all at once.
   They are memory-efficient and are useful for large data or infinite sequences.

    Key Characteristics:
     * Use the **yield** keyword instead of return.
     * Maintain their state between calls.
     * Automatically create iterators.


       def countdown(n):
           while n > 0:
           yield n
           n -= 1
            
            # Create generator
            gen = countdown(3)
            # Iterate through it
            for number in gen:
                print(number)


 7. What are the advantages of using generators over regular functions?

   Ans:

    1. Memory Efficient

       Generators produce one item at a time and don't store the entire result in memory.
       
       Regular functions return entire data structures (like lists), which can consume lots of memory.

    2. Lazy Evaluation
     
       Generators compute values only when needed (on the fly).
       
       Great for infinite sequences or when processing may not need all items.

    3. Faster Startup Time
     
        Generators start yielding values immediately.
        
        Regular functions may take time to compute and return the full result first.

        
         # Generator
          def count_up_to(n):
              count = 1
              while count <= n:
                   yield count
                   count += 1
                   
         # Regular function
            def count_up_to_list(n):
              return [i for i in range(1, n+1)]


 8. What is a lambda function in Python and when is it typically used?

   Ans:

   In Python, a lambda function is a small, anonymous function defined using the lambda keyword. It can take any number of arguments but can only have one expression. Lambda functions are typically used for short, concise operations, especially when passing functions as arguments to higher-order functions like map, filter, or reduce.

   Key Characteristics of Lambda Functions:
    
  Anonymous: They don't have a name, unlike regular functions defined with def.

  Single Expression: Lambda functions can only contain a single expression, which is automatically returned.
  
  Concise: They provide a way to create small, throwaway functions quickly, without needing a full function definition.  
   

      # Regular function
         def square(x):
             return x * x
             
      # Equivalent lambda function
         square_lambda = lambda x: x * x
         print(square(5))        # Output: 25
         print(square_lambda(5)) # Output: 25

 10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

   Ans:
   1. map(function, iterable)
      Purpose: Applies a function to every item of an iterable.

      Returns: A new iterable with the transformed values.

      E.g
           
              nums = [1, 2, 3]   
              squared = list(map(lambda x: x * x, nums))
              # Output: [1, 4, 9]

    2. filter(function, iterable)
       Purpose: Applies a function to filter items (only those that return True).
       
       Returns: A new iterable with only the items that satisfy the condition.

       E.g

               nums = [1, 2, 3, 4, 5]
               even = list(filter(lambda x: x % 2 == 0, nums))
               # Output: [2, 4]

     3. reduce(function, iterable)
        Purpose: Applies a function to accumulate or combine items, reducing them to a single value.

        Returns: A single result.
        
        Needs to be imported from functools.
        
        E.g

              from functools import reduce
              nums = [1, 2, 3, 4]
              product = reduce(lambda x, y: x * y, nums)
              # Output: 24

```python
#Q2. Create a Python function that accepts a string and returns the reverse of that string
def reverse_string(my_str):
    reversed_str = ''
    for char in my_str:
        reversed_str = char + reversed_str
    return reversed_str

my_str = input("Enter the string: ")
print("Reversed string:", reverse_string(my_str))
```

```python
#Q4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, int(num**0.5)+1):
        if num % i == 0:
            return False
    return True

# Check for numbers from 1 to 200
for number in range(1, 201):
    if is_prime(number):
        print(number, end=" ")
```

```python
#Q5.  Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.count = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.b

# Example usage:
n = int(input("Enter number of terms: "))
fib = FibonacciIterator(n)
for num in fib:
    print(num, end=' ')
```

```python
#Q6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def powers_of_two(max_exponent):
    for exp in range(max_exponent + 1):
        yield 2 ** exp
n = int(input("Enter the maximum exponent: "))
for power in powers_of_two(n):
    print(power, end=' ')
```

```python
#Q7.  Implement a generator function that reads a file line by line and yields each line as a string.
with open("example.txt","w") as f:
  f.write("Hello\n")
  f.write("Varsha")

def read_file(filename):
  with open(filename,'r') as file:
    for line in file:
      yield line.strip()

for line in read_file("example.txt"):
  print(line)
```

```python
#Q9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
# List of temperatures in Celsius
celsius_temps = [0, 20, 37, 100, -10]

# Use map with lambda to convert to Fahrenheit
fahrenheit_temps = list(map(lambda c: (c * 9/5) + 32, celsius_temps))

# Output the result
print("Celsius temperatures:", celsius_temps)
print("Fahrenheit temperatures:", fahrenheit_temps)
```

```python
#Q10. Create a Python program that uses `filter()` to remove all the vowels from a given string.
def remove_vowels(input_str):
    vowels = "aeiouAEIOU"
    return ''.join(filter(lambda char: char not in vowels, input_str))

# Get input from user
user_input = input("Enter a string: ")

# Remove vowels using filter
result = remove_vowels(user_input)

# Output the result
print("String without vowels:", result)
```
